```python
# Importing modules
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import cv2 # OpenCV
import glob # loading paths

# load all images (pngs)
images = glob.glob('*.jpg') #** means all the pictures of the array, and * means only 1 picture

print(images)

print(type(images))
print(np.shape(images))
print(images[1])
```

```python
# load as color image
img = cv2.imread(images[1], cv2.IMREAD_COLOR)

# load as grayscale
img_gray = cv2.imread(images[1], cv2.IMREAD_GRAYSCALE)
print()

#print(np.shape(img))
#print(img)

#print(np.shape(img_gray))
#print(img_gray)
#z = np.ones([2,2,5], dtype=bool)
#print(z)
```

```python
# load color image with integers
img = cv2.imread(images[1], 1)

# grayscale
img_gray = cv2.imread(images[1], 0)
```

```python
# display color and grayscale images
fig, axes = plt.subplots(1, 2, figsize=(4, 4), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].set_title("Original color image")
ax[0].imshow(img[...,::-1], interpolation = "nearest")

ax[1].set_title("Image as grayscale")
ax[1].imshow(img_gray, cmap=plt.cm.Greys_r, interpolation = "nearest")

for a in ax:
    a.set_axis_off()

fig.tight_layout()
plt.show()
```

```python
center = (width/2, height/2)

# using cv2.getRotationMatrix2D() to get the rotation matrix
# angle is rotating angle, scale is zooming in or out (from where the center) 
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=45, scale=1)
```

```python
# rotate the image using cv2.warpAffine
rotated_image = cv2.warpAffine(src=img, M=rotate_matrix, dsize=(width, height))
```

```python
# cv2.imshow('Original image', img)
cv2.imshow('Rotated image', rotated_image)
# wait indefinitely, press any key on keyboard to exit
cv2.waitKey(0)
# save the rotated image to disk
cv2.imwrite('rotated_image.jpg', rotated_image)
```
