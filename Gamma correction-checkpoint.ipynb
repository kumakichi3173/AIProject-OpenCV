```python
# Importing modules
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import cv2 # OpenCV
import glob # loading paths

# load an image (in png)
images = glob.glob('*.png') #** means all the pictures of the array, and * means only 1 image

print(images)

print(type(images))
print(np.shape(images))
print(images[2])
```

```python
# load as color image
# [0] is the aerial image in the array and [1] is the fracture_spine (MRI) image in the array
# see what pictures are in the array above result
img = cv2.imread(images[2], cv2.IMREAD_COLOR) 
print()
```

```python
# apply gamma correction on the MRI image
# displaying the title
plt.title("Original MRI Image")

plt.imshow(img)
```

```python
def adjust_gamma(image, gamma=1.0):
    # build a lookup table mapping the pixel values [0, 255] to
    # their adjusted gamma values
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")

    # apply gamma correction using the lookup table
    return cv2.LUT(image, table)
```

```python
adjusted = adjust_gamma(img, gamma=2.35)

# print(adjusted)
# (row, column, channels like RGB)
print(np.shape(adjusted))
```

```python
# cv2.imshow('Original image', img)
cv2.imshow('adjusted', adjusted)
# wait indefinitely, press any key on keyboard to exit
cv2.waitKey(0)
# save the rotated image to disk
cv2.imwrite('gamma_adjusted.jpg', adjusted)
```

```python
# display original and gamma adjusted images
fig, axes = plt.subplots(1, 2, figsize=(10, 10), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].set_title("Original MRI image")
ax[0].imshow(img)

ax[1].set_title("Gamma Adjusted Image")
ax[1].imshow(adjusted, cmap=plt.cm.Greys_r, interpolation = "nearest")

for a in ax:
    a.set_axis_off()

fig.tight_layout()
plt.show()
```
