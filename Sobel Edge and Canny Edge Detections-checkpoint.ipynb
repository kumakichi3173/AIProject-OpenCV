```python
# This program is about a Sobel Edge Detection and a Canny Edge Detection 
# I tried with 2 different images to see how Sobel Filter works on X-axis and Y-axis 
# # The Sobel X image is filtered X axis out and the Sobel Y image is filtered Y axis out 
# Compare the image to a Canny Edge Detection and see if there is a difference
# In addition, I also learned how to use this code on Google Colab which has the same extension (.ipynb)
```

```python
# Importing modules
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import cv2 # OpenCV
import glob # loading paths

# load an image (in png)
argv = glob.glob('*.png') #** means all the pictures of the array, and * means only 1 image

print(argv)

print(type(argv))
print(np.shape(argv)) # how many images are in the directory
print(argv[4]) # you want to use the 4th image 'SobelEdgeTiger.png' in the list
```

```python
src = cv2.imread('SobelEdgeTiger.png')
# OpenCV reads images in BGR, but matplotlib shows images in RGB, so
# you have to convert to RGB before using plt.imshow
plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))

img = cv2.imread(argv[4], cv2.IMREAD_COLOR) 
print()
```

```python
# If you don't have the following code, 
# "plt.imshow(gray, cmap='gray'),"
# matplotlib reads the grayscale image into a blue green color by default
bgdefault = cv2.imread('SobelEdgeTiger.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(bgdefault)
```

```python
# Now, channge the image into grayscale 

# either codes work:

#gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
#plt.imshow(gray) # show the image
# OR

gray = cv2.imread('SobelEdgeTiger.png', cv2.IMREAD_GRAYSCALE)
# cmap needs to be set as gray because matplotlib reads the grayscale image into a blue green color by default
# when you want to use plt package, you have to indicate "...., cmap='gray')"
plt.imshow(gray, cmap='gray') # show the image
```

```python
# Sobel Filter
scale = 1
delta = 0
ddepth = cv2.CV_16S

grad_x = cv2.Sobel(gray, ddepth, 1, 0, ksize=3, scale=scale, delta=delta, borderType=cv2.BORDER_DEFAULT)
grad_y = cv2.Sobel(gray, ddepth, 0, 1, ksize=3, scale=scale, delta=delta, borderType=cv2.BORDER_DEFAULT)

#Conversion Filter
abs_grad_x = cv2.convertScaleAbs(grad_x)
abs_grad_y = cv2.convertScaleAbs(grad_y)
grad = cv2.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)
```

```python
# Sobel Filter only on x axis
# The Sobel X image is filtered X axis out 
plt.imshow(abs_grad_x, cmap='gray')
```

```python
# Sobel Filter only on y axis
# The Sobel Y image is filtered X axis out  
plt.imshow(abs_grad_y, cmap='gray')
```

```python
#Conversion Filter
abs_grad_x = cv2.convertScaleAbs(grad_x)
abs_grad_y = cv2.convertScaleAbs(grad_y)

grad = cv2.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)
```

```python
##Display the image after conversion filter
plt.imshow(grad, cmap='gray')
```

```python
gray = cv2.imread('SobelXY.png', cv2.IMREAD_GRAYSCALE)
# cmap needs to be set as gray because matplotlib reads the grayscale image into a blue green color by default
# when you want to use plt package, you have to indicate "...., cmap='gray')"
plt.imshow(gray, cmap='gray') # show the image
```

```python
# Sobel Filter
scale = 1
delta = 0
ddepth = cv2.CV_16S

grad_x = cv2.Sobel(gray, ddepth, 1, 0, ksize=3, scale=scale, delta=delta, borderType=cv2.BORDER_DEFAULT)
grad_y = cv2.Sobel(gray, ddepth, 0, 1, ksize=3, scale=scale, delta=delta, borderType=cv2.BORDER_DEFAULT)

#Conversion Filter
abs_grad_x = cv2.convertScaleAbs(grad_x)
abs_grad_y = cv2.convertScaleAbs(grad_y)
grad = cv2.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)
```

```python
#Display the image after conversion filter
plt.imshow(grad, cmap='gray')
```
